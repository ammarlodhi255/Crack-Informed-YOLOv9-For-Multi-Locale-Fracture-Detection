```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the table
categories = ['mAP@50', 'Precision', 'Sensitivity']
benchmark = [0.56, 0.81, 0.47]
ours = [0.59, 0.86, 0.53]

x = np.arange(len(categories))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(8, 5))
rects1 = ax.bar(x - width/2, benchmark, width, label="Author's Benchmark", color='tab:blue')
rects2 = ax.bar(x + width/2, ours, width, label='Ours', color='tab:red')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
# ax.set_title('Comparison of FracAtlas Benchmark with Our Approach on the Validation Set')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()

# Function to add labels on the bars
def add_labels(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

add_labels(rects1)
add_labels(rects2)

fig.tight_layout()

# Save the figure
fig.savefig('FracAtlas_Comparison.png')

plt.show()
```
